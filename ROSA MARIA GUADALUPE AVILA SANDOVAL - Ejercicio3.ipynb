**Ejercicio 1**

Complete el código de la calculadora en extenso.

```python
#Ejemplo en extenso del menú de una calculadora sencilla

opc = input("¡Bienvenido a la calculadora!\n\nA. SUMA \nB. RESTA \nC. MULTIPLICACIÓN \nD. DIVISIÓN \n\nSeleccione una opción:")#\n salto de línea

if opc == 'A' or opc == 'a': # Suma
  D1 = float(input("\nIngrese el primer número: "))
  D2 = float(input("Ingrese el segundo número: "))
  print("\nEl resultado de la suma es: {:.2f}".format(D1+D2))

elif opc == 'B' or opc == 'b': # Resta
  D1 = float(input("\nIngrese el primer número: "))
  D2 = float(input("Ingrese el segundo número: "))
  print("\nEl resultado de la resta es: {:.2f}".format(D1-D2))

elif opc == 'C' or opc == 'c' : #multiplicación
  D1 = float(input("\nIngrese el primer número: "))
  D2 = float(input("Ingrese el segundo número: "))
  print("\nEl resultado de la multiplicación es: {:.2f}".format(D1*D2))

elif opc == 'D' or opc == 'd': # División
  D1 = float(input("\nIngrese el primer número: "))
  D2 = float(input("Ingrese el segundo número: "))
  print("\nEl resultado de la división es: {:.2f}".format(D1/D2))

else:
  print("\nDato no válido")

  
```

**Ejercicio 2**

Complete el código de la calculadora resumida.

```python
#Ejemplo resumido del menú de una calculadora sencilla

opc = input("¡Bienvenido a la calculadora! \n\nA. SUMA  \nB. RESTA \nC. MULTIPLICACIÓN \nD. DIVISIÓN \n\nSeleccione una opción:")#\n salto de línea
opc = opc.upper() # upper() Instrucción que transforma a mayúsculas lower() Instrucción que transforma a minúsculas
D1 = float(input("\nIngrese el primer número: "))
D2 = float(input("Ingrese el segundo número: "))

if opc == 'A': #Suma
  print("\nEl resultado de la suma es: {:.2f}".format(D1+D2))

elif opc == 'B': #Resta
  print("\nEl resultado de la resta es: {:.2f}".format(D1-D2))

elif opc == 'C': #Multiplicación
  print("\nEl resultado de la multiplicación es: {:.2f}".format(D1*D2)) 

elif opc == 'D': #División
  print("\nEl resultado de la división es: {:.2f}".format(D1/D2))

else:
  print("\nDato no válido")

```
