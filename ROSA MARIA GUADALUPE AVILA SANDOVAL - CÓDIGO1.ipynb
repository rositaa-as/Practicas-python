```python
# para comentarios uso del gato

#Variables
A = 9+3*16/2**3-5/2
B = 6+2*(7-(4+2))-3*(10+3**2/6)
C = 9%5 + (not (5*4%3*7<33 or 27/2*3<=10*3/10))

#print() para imprimir datos

print("Resultado del ejemplo 1:", A)
print("Resultado del ejemplo 2:", B)
print("resultado del ejemplo 3:", C)
```
