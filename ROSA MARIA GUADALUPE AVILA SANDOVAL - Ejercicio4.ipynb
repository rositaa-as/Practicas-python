# **Ejemplo 3**

---

> ‚ÄúLa langosta ahumada‚Äù es una empresa dedicada a ofrecer banquetes; sus tarifas son las siguientes: el costo de platillo por persona es de 95.00, pero si el n√∫mero de personas es mayor a 200 pero menor o igual a 300, el costo es de 85.00. Para m√°s de 300 personas el costo por platillo es de 75.00. Se requiere un algoritmo que ayude a determinar el presupuesto que se debe presentar a los clientes que deseen realizar un evento.



# **Variables de entrada y salida**
image.png
# **Pseudoc√≥digo**


```
1. Inicio
2. Leer NP
3. Si NP > 300
      Hacer TOT = NP * 75 
   Si no
      Si NP > 200
          Hacer TOT = NP * 85 
      Si no
          Hacer TOT = NP * 95 
4. Escribir ‚ÄúEl total es‚Äù, TOT
5. Fin
```


# **Diagrama de Flujo**
image.png

````
#C√≥digo Ejemplo 3: Presupuesto Banquete

NP = int(input("Ingrese el n√∫mero total de personas: "))

if NP > 300:
  TOT = NP*75.00
else:
  if NP > 200:
    TOT = NP * 85.00
  else:
    TOT = NP * 95.00

print("El precio total por {} personas es: ${} pesos".format(NP, TOT))
````
