**Ejercicio 1 -** Operaciones sencillas

```python
print("--INICIO--\n")
a= float(input("Ingresa el primer número: "))
b= float(input("Ingresa el segundo número: "))
print("Suma: ",a+b)

#mostrar 2 decimales
print("Resta:",a-b)
print("Resta: {:.2f} \n".format(a-b))


print("Multiplicación: ",a*b)
print("la multiplicación de {} más {} es {:.2f}".format(a,b,a*b))
print("la multiplicación de", a ,"más" ,b, "es", a*b)
print("\nDivisión: ",b/a)
print("--FIN--")
```
