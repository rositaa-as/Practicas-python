> Se requiere un algoritmo para obtener la suma de diez cantidades mediante la utilización de un ciclo “Mientras”.

**Variables de entrada y salida**

image.png

**Pseudocódigo**


```
1. Inicio
2. Hacer SU = 0
3. Hacer C = 1
4. Mientras C < = 10
      Leer VA
      Hacer SU = SU + VA
      Hacer C = C + 1
   Fin mientras 
5. Escribir SU
6. Fin
```


**Diagrama de Flujo**

image.png

````python
#Código Ejemplo 1

SU = 0 # Variable acumuladora
C = 1 # Variable de tipo contador

while C <= 10:
  VA = float(input("Ingrese un valor: "))
  SU = SU + VA
  C = C + 1

print("La suma total de todos los números ingresados es: ", SU)
````

> Se requiere un algoritmo para obtener la edad promedio de un grupo de N alumnos.


**Variables de entrada y salida**

chrome_GLCCsanrxU.png

**Pseudocódigo**

```
1. Inicio
2. Hacer SU = 0
3. Leer NU
4. Hacer C = 1
5. Mientras C < = NU
      Leer ED
      Hacer SU = SU + ED
      Hacer C = C + 1
   Fin mientras 
6. Hacer PR = SU / NU
7. Escribir PR
8. Fin
```

**Diagrama de Flujo**

image.png


````python
#Código Ejemplo 2

SU = 0
NU = int(input("Ingrese el número de alumnos:"))
C = 1

while C <= NU:
  ED = int(input("Ingrese la edad del alumno {}:".format(C)))
  SU = SU + ED
  C = C + 1

PR = SU/NU
print("La edad promedio del grupo es: ", PR)
````

# **EJEMPLO 3**

---


> Se requiere un algoritmo para determinar cuánto ahorrará una persona en un año, si al final de cada mes deposita variables cantidades de dinero; además, se requiere saber cuánto lleva ahorrado cada mes


**Variables de entrada y salida**

image.png

**Pseudocódigo**

```
1. Inicio
2. Hacer AH = 0
3. Hacer M = 1
4. Mientras M < = 12
      Leer CA
      Hacer AH = AH + CA
      Hacer M = M + 1
      Escribir “El ahorro del mes:”, M, “es”, AH
   Fin mientras 
5. Escribir “El ahorro final es:”, AH
6. Fin
```

**Diagrama de Flujo**

image.png

````python
#Código Ejemplo 3

AH, M = 0, 1

while M <= 12:
  CA = float(input("Ingrese la cantidad que ahorro en el mes {}: ".format(M)))
  AH = AH + CA # AH += CA
  print("Su ahorro para el mes {} es de: {}".format(M,AH))
  M += 1 # M = M + 1

print("Su ahorro total de un año es: ", AH)
````
