**Ejercicio 1**

> Solicite al usuario que ingrese una frase e imprima cuantas veces aparece cada vocal (a, e, i, o, u).

**Ejemplo de salida en pantalla:**


Ingresa una frase: Hola a todos

La letra a aparece 2 veces

La letra e aparece 0 veces

La letra i aparece 0 veces

La letra o aparece 3 veces

La letra u aparece 0 veces


```python
print("Bienvenido al contador de vocales.")
frase= str(input("\nIngrese una frase: "))

#contadores
ac=0
ec=0
ic=0
oc=0
uc=0

for letra in frase:
  if letra == "a" or letra == "A":
    ac += 1
  elif letra == "e" or letra == "E":
    ec += 1
  elif letra == "i" or letra == "I":
    ic += 1
  elif letra == "o" or letra == "O":
    oc += 1
  elif letra == "u" or letra == "U":
    uc +=1
print("La letra a aparece {} veces. \nLa letra e aparece {} veces. \nLa letra i aparece {} veces. \nLa letra o aparece {} veces. \nLa letra u aparece {} veces.".format(ac,ec, ic, oc,uc))


```

> Muestre en pantalla todos los números impares que se encuentran del 1 al 80.

> Realice un algoritmo para obtener la tabla de multiplicar de un en-
tero K comenzando desde el 1.



```python
tabla = int(input("Ingrese un número para saber su tabla de multiplicar: "))
resu=0

for numero in range(1,11):
  resu= tabla*numero
  print("{} por {} es: {}. ".format(tabla, numero, resu))


```
