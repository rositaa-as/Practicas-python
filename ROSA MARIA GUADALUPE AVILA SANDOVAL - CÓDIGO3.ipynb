Ejemplo 1: sentencia **if**

```python
print("------INICIO-------\n")

nota= int(input("Ingresa tu calificación: "))

if nota>=60:
  print("\n¡Felicidades aprobaste el examen!")

  print("\n-----FIN------")
```

Ejemplo 2: sentencia **if-else**


```python
print("------INICIO-------\n")

nota= int(input("Ingresa tu calificación: "))

if nota>=60:
  print("\n¡Felicidades aprobaste el examen!")

else:
  print("\n¡Animo! la proxima vez lo harás mejor :)")

  print("\n-----FIN------")
```

Ejemplo 3: sentencia **if-elif-else**

```python
print("\n-----INICIO------")

nota = int(input("Ingresa tu calificación: "))

if nota>=80:
  print("\n¡Felicidades, aprovaste el examen!")

elif nota>=60:
    print("\n¡Buen trabajo! pero deberás esforzarte más")

else:
   print("\n¡Ánimo! la proxima vez lo harás mejor")

print("\n-----FIN------")
```

Establecer limites 

```python
print("\n-----INICIO------")

nota = int(input("Ingresa tu calificación: "))

if nota<0 or nota>100:
  print("\nEl dato no es válido, ingrese una calificación de 0 a 100")

elif nota>=80:
  print("\n¡Felicidades, aprovaste el examen!")

elif nota>=60:
    print("\n¡Buen trabajo! pero deberás esforzarte más")

else:
   print("\n¡Ánimo! la proxima vez lo harás mejor")

print("\n-----FIN------")
```
