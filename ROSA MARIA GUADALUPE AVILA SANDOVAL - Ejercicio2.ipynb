
````
#Código Ejemplo 1 Promedio de 4 calificaciones

#Lectura y asignación de las 4 calificaciones
C1 = float(input("Ingrese su primera calificación: "))
C2 = float(input("Ingrese su segunda calificación: "))
C3 = float(input("Ingrese su tercera calificación: "))
C4 = float(input("Ingrese su cuarta calificación: "))

#Cálculo del promedio de las 4 calificaciones
S = C1 + C2 + C3+ C4
P = S / 4

#Despliegue de resultados
print("\nSu promedio es: {:.2F}".format(P))
````

# **Ejemplo 2**

Se requiere determinar el **sueldo semanal** de un trabajador con base en las **horas que trabaja** y el **pago por hora** que recibe.

# **Variables de entrada y salida**

image.png

# **Pseudocódigo**


```
1.- Inicio
2.- Leer HT, PH

```

# **Ejemplo 3**

---

La conagua requiere determinar el pago que debe realizar una persona por el total de metros cúbicos que consume de agua al llenar una alberca 
(ver figura). Realice un algoritmo que permita determinar ese pago.

image.png


# **Variables de entrada y salida**

image.png

# **Pseudocódigo**


```
1.- Inicio
2.- Leer A, L, N, CM
3.- Hacer V = (A*L*N)
4.- Hacer PAG = V * CM
5.- Escribir PAG
6.- Fin
```



# **Diagrama de flujo**

image.png

````
#Código Ejemplo 3 Pago a Conagua

#Lectura y asignación de la altura, largo y ancho de la alberca asi como el costo por m3
A = float(input("Ingrese la altura de la alberca en m: "))
L = float(input("Ingrese el largo de la alberca en m: "))
N = float(input("Ingrese el ancho de la alberca en m: "))
CM = float(input("Ingrese el costo por m3 de agua: "))

#Cálculo del pago a Conagua
V = A * L * N
PAG = V * CM

#Despliegue de resultados
print("\nEl pago a realizar por {}m3 de agua es: ${:.2F} pesos".format(V, PAG))
````

# **Ejercicio 1**

---




> Una papelería vende plumas a 10.50, cuadernos a 46, y caja de colores a 18.75. Calcule el monto total de una venta según el número de artículos solicitados por el usuario. Muestre el total de la venta en pantalla.



````
#Venta
P = float(input("Ingrese la cantidad de plumas necesita: "))
Cu = float(input("Ingrese la cantidad de cuadernos que necesita: "))
CC = float(input("Ingrese la cantidad de caja de colores que necesita: "))

A = P * 10.50
B = Cu * 46
C = CC * 18.75

Pago = A+B+C

print("\nEl total a pagar es: ${:.2f}".format(Pago))
````

# **Ejercicio 3**

---



>  Una empresa desea determinar el monto de un cheque que debe proporcionar a uno de sus empleados que tendrá que ir por cierto número de días a la ciudad de Monterrey; los gastos que cubre la empresa son: hotel, comida y 100.00 pesos diarios para otros gastos. Realice un algoritmo que determine el monto total del cheque y muestre tambien el desglose por cada concepto.


````
#monto del cheque

Dia = float(input("ingrese los días de viaje: "))
Hotel = float(input("Ingrese el costo del Hotel por dia: "))
Comida = float(input("ingrese el gasto de las comidas por dia: "))

Extra = 100

GH = Dia * Hotel
GC = Dia * Comida
GE = Dia * Extra

Cheque = GH+GC+GE

print("\nLos gastos del viaje de", Dia, "días es de:" )
print("\nGasto del hotel en total es de: ${:.2f}".format(GH))
print("Gasto de la comida en total es de: ${:.2f}".format(GC))
print("Gasto de los $100 extras en total es de: ${:.2f}".format(GE))
print("\nMonto total del cheque es de: ${:.2f}".format(Cheque))
````
